Data cleaning

```python
# Menghapus duplicated
df = df.drop_duplicates()

df.duplicated().sum()

df.head()
```

```python
# Rename column
df = df.rename(columns={"cp": "chest_pain_type"})
df = df.rename(columns={"trestbps" : "resting_blood_pressure"})
df = df.rename(columns={"col": "serum_cholesterol"})
df = df.rename(columns={"fbs": "fasting_blood_sugar"})
df = df.rename(columns={"thal": "thalassemia"})
df.head()
```

```python
# Mengubah semua column menjadi capitalize
df.columns = df.columns.str.capitalize()
df.head()
```
